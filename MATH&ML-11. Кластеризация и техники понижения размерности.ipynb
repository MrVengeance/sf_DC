```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import \
    classification_report, roc_auc_score, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score
from sklearn import metrics

from sklearn.metrics.cluster import homogeneity_score
from sklearn.metrics.cluster import completeness_score
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
df = pd.read_csv('data/Country-data.csv')
df.head()
```

```python
df['life_expec'].max()
```

```python
df_new=df.drop(['country'], axis= 1)
```

```python
X=df_new.values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() 
scaler.fit(X) 
X_norm = scaler.transform(X) 
X_norm[0][0]
```

```python
df_new.corr()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.9, random_state=42).fit(X_norm)
len(pca.components_)
```

```python
pca.explained_variance_ratio_[0]
```

```python

import seaborn as sns
```

```python
X_pca = pca.fit_transform(X_norm)
corrmat = np.corrcoef(X_pca.transpose())
sns.heatmap(corrmat, annot=True)
```

```python
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for i in range_n_clusters:
    kmeans = KMeans(n_clusters=i, random_state = 1)
    kmeans.fit(X_pca)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(X_pca, cluster_labels)
    print(i, silhouette_avg)
```

```python

from matplotlib import pyplot as plt
```

```python
kmeans = KMeans(n_clusters=4, random_state=1)
kmeans.fit(X_pca)
df['K-Means_Cluster_ID'] = kmeans.labels_
plt.figure(figsize=(8,4),dpi=200)
sns.boxplot(x='K-Means_Cluster_ID', y='gdpp', data=df)
plt.figure(figsize=(8,4),dpi=200)
sns.boxplot(x='K-Means_Cluster_ID', y='child_mort', data=df)
```
